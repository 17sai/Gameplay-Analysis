# outlier detection

```python
def print_no_of_outliers(x):
    q1 = a[x].quantile(0.25)
    q3 = a[x].quantile(0.75)
    iqr = q3 - q1
    lb = q1 - 1.5 * iqr
    ub = q3 + 1.5 * iqr
    length = len(a[(a[x] < lb) | (a[x] > ub)])
    print(length)
```

# outlier replacement

```python
def cap_outliers_iqr(df,column):
    Q1 = np.percentile(df[column].dropna(), 25)
    Q3 = np.percentile(df[column].dropna(), 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df.loc[df[column] < lower_bound, column] = lower_bound
    df.loc[df[column] > upper_bound, column] = upper_bound
    return df
```

```python
cap_outliers_iqr(a,"assists")
```

```python
cap_outliers_iqr(a,"boosts")
```

```python
cap_outliers_iqr(a,"damageDealt")
```
