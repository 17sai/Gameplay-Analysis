# Univariate Analysis

## Categorical data

```python
# 1. kill_style â†’ Pie Chart
plt.figure(figsize=(6,6))
a['kill_style'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette("Set2"))
plt.title("Distribution of Kill Styles")
plt.ylabel("")
plt.show()
```

```python
# 2. match_category â†’ Countplot
plt.figure(figsize=(7,4))
sns.countplot(data=a, x="match_category", order=a["match_category"].value_counts().index, palette="viridis")
plt.title("Match Category Distribution")
plt.show()
```

```python
# 3. healing_style â†’ Bar Chart
plt.figure(figsize=(7,4))
a['healing_style'].value_counts().plot(kind='bar', color="teal")
plt.title("Healing Style Frequency")
plt.ylabel("Count")
plt.show()
```

```python
# 4. distance_type â†’ Countplot
plt.figure(figsize=(7,4))
sns.countplot(data=a, x="distance_type", palette="muted")
plt.title("Distance Type Distribution")
plt.show()
```

```python
# 5. performance_tag â†’ Pie Chart
plt.figure(figsize=(6,6))
a['performance_tag'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette("coolwarm"))
plt.title("Performance Tag Distribution")
plt.ylabel("")
plt.show()
```

## Numerical data

```python
# 6. kills â†’ Histogram + KDE
plt.figure(figsize=(7,4))
sns.histplot(a['kills'], bins=30, kde=True, color="skyblue")
plt.title("Kills Distribution")
plt.show()
```

```python
# 9. swimDistance â†’ Histogram
plt.figure(figsize=(7,4))
sns.histplot(a['swimDistance'], bins=30, color="green", kde=False)
plt.title("Swim Distance Distribution")
plt.show()
```

```python
# 10. assists â†’ Countplot
plt.figure(figsize=(7,4))
sns.countplot(data=a, x="assists", palette="Set1")
plt.title("Assists Count Distribution")
plt.show()
```

```python
# 12. winPlacePerc â†’ KDE Plot
plt.figure(figsize=(7,4))
sns.kdeplot(a['winPlacePerc'], fill=True, color="blue", alpha=0.4)
plt.title("Win Percentage Distribution")
plt.xlabel("Win Place Percentage")
plt.show()
```

# Bivariate Analysis

```python
# 1. Categorical vs Categorical â†’ Countplot
plt.figure(figsize=(8,5))
sns.countplot(data=a, x="match_category", hue="distance_type")
plt.title("Match Category vs Distance Type")
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Categorical vs Numerical â†’ Boxplot
plt.figure(figsize=(8,5))
sns.boxplot(data=a, x="match_category", y="kills")
plt.title("Kills Distribution per Match Category")
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Categorical vs Numerical â†’ Barplot (Mean values)
plt.figure(figsize=(8,5))
sns.barplot(data=a, x="healing_style", y="damageDealt", estimator="mean")
plt.title("Average Damage Dealt by Healing Style")
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Numerical vs Numerical â†’ Scatterplot
plt.figure(figsize=(8,5))
sns.scatterplot(data=a, x="walkDistance", y="winPlacePerc", alpha=0.3)
plt.title("Walk Distance vs Win Percentage")
plt.show()
```

```python
# 5. Numerical vs Numerical â†’ Correlation Heatmap
num_cols = ["kills","damageDealt","walkDistance","boosts","weaponsAcquired","winPlacePerc"]

corr_matrix = a[num_cols].corr()

plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Numerical Features")
plt.show()
```

# Multivariate Analysis

```python
# 1. Pairplot (All Numerical Variables Together)
num_cols = ["kills","damageDealt","walkDistance","swimDistance",
            "assists","revives","boosts","weaponsAcquired","winPlacePerc"]

sns.pairplot(a[num_cols], diag_kind="kde", corner=True, plot_kws={'alpha':0.3})
plt.suptitle("Pairwise Relationships Between Numerical Features", y=1.02)
plt.show()
```
